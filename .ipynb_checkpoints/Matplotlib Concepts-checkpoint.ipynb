```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **9 + x ** 5 + x **4 + x**3 + x ** 2 + x +1
```

```python
# Functional Method
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
# Multiplots on the same convas
plt.subplot(1,2,1)
plt.plot(x, y, "r")
plt.subplot(1,2,2)
plt.plot(x, y **(-1), "b")
```

```python
# Object Oriented Method
fig = plt.figure()

fig_axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

fig_axes.plot(x, y)
x_label = "X Label"
fig_axes.set_xlabel(x_label)
Y_label = "Y Label"
fig_axes.set_ylabel(Y_label)
chart_title = "Chart Title"
fig_axes.set_title(chart_title)
```

```python
# Object Oriented Method
fig = plt.figure()

fig_1 = fig.add_axes([0,0,1,1])
fig_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

fig_1.plot(x,y)
fig_1.set_title("Poly")

fig_2.plot(y,x)
fig_2.set_title("Reverse Poly")
```
