```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = x ** 2
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(5,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (5,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig("my_fig.png")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label="X2")
ax.plot(x, x**3, label = "X3")
ax.plot(x, np.log(x**4), label = "X4")
ax.legend(loc=0)
```
